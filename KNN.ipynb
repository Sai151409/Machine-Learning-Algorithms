```python
def outliers(df, columns):
    for i in columns:
        q1 = np.quantile(df[i], 0.25)
        q3 = np.quantile(df[i], 0.75)
        iqr = q3 - q1
        high = q3 + 1.5 * iqr
        low = q1 - 1.5 * iqr
        print((len(df[(df[i] < low) & (df[i] > high)])))
```

```python
outliers(df1, df1.columns[:-1])
```

```python
data = {}
for i in df1.columns[:-1]:
    data[i] = {}
    for j in df1['TARGET CLASS'].unique():
        data[i][j] = df1[df1['TARGET CLASS'] == j][i]
```

```python
data.keys()
```

```python
data['WTT'].keys()
```

```python
from scipy.stats import f_oneway

for i in data.keys():
    _, p_value = f_oneway(data[i][1], data[i][0])
    print(f"P-value of {i} : {p_value}")
    if p_value > 0.05:
        print(f"There is no relationship between means of {list(data[i].keys())}")
    else:
        print(f"There is relationship between means of {list(data[i].keys())}")
    print("\n")
```

```python
from  sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df1.iloc[:, :-1], df1.iloc[:, -1], test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
X_train = scale.fit_transform(X_train)
X_test = scale.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train,y_train)
pred = knn.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.model_selection import cross_val_score
print(confusion_matrix(y_test,pred))
```

```python
print(classification_report(y_test,pred))
```

```python
accuracy_rate = []
for i in range(1, 40):
    knn = KNeighborsClassifier(n_neighbors=i)
    score = cross_val_score(knn, X_train, y_train, cv = 10)
    accuracy_rate.append(np.mean(score))
```

```python
max(accuracy_rate)
```

```python
accuracy_rate
```

```python
train_error_rate = []
test_error_rate = []
for i in range(1, 40):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    train_pred_i = knn.predict(X_train)
    train_error_rate.append(np.mean(train_pred_i != y_train))
    test_pred_i = knn.predict(X_test)
    test_error_rate.append(np.mean(test_pred_i != y_test))
```
